```python
import pandas as pd
import folium
import folium.plugins

import requests #Criando mapa

dados_geo = pd.read_json("br.json")


estado = 'RJ'
url = "https://servicodados.ibge.gov.br/api/v3/malhas/estados/{}?formato=application/vnd.geo+json".format(estado)
response = requests.get(url, verify=False)
geo_json_data = requests.get(url).json()

style =  {'fillColor': 'blue', #cor de preenchimento
            'color': 'red',#cor da linha de contorno
            'weight': 0.0, #espessura da linha
            }
#OpÃ§Ã£o


min_lon, max_lon = -45, -35
min_lat, max_lat = -25, -10

mapa3 = folium.Map(
    max_bounds=True,
    location=[-15.793889, -47.882778],
    zoom_start=4,
    min_lat=min_lat,
    max_lat=max_lat,
    min_lon=min_lon,
    max_lon=max_lon,)

folium.plugins.Geocoder().add_to(mapa3)

folium.GeoJson(geo_json_data, style_function=lambda x:style).add_to(mapa3)

cidade =  dados_geo["city"]
latitude =  dados_geo["lat"]
longitude = dados_geo["lng"]

for c, lat, lon in zip(cidade, latitude, longitude):
    folium.Marker([lat,lon], tooltip=c).add_to(mapa3)

mapa3
```
